```python
import numpy as np
import random as random

def column_stochastic_inator(M):

    M = M.astype(float)
    n = M.shape[1] #n is the number of columns, which in this case is websites

    #fix all-zero rows or "dead ends"
    for i in range(n):
        if np.sum(M[:, i]) == 0:
            M[:, i] = 1/n

    M = M/np.sum(M, axis=0) # makes all non-zero rows "normalized" or sum to one (column stochastic)

    return M

def Google_matrix_inator(G, x=.85):
    
    G = G.astype(float)
    a = G.shape[1] #a is the number of columns, which in this case is websites

    J = np.ones((a, a)) / a #jump matrix for initial markov chain
    
    G = x * G + (1 - x) * J #damping formula
    
    return G

def power_method_inator(G, i): #this strecthes all eignvalues and then renormalizes
    
    n = G.shape[0]

    v = np.ones(n)/n

    for i in range(i):
        new_v = (G @ v)
        
        new_v = new_v/np.sum(new_v)

        v = new_v

    return v
```

```python
def PageRank(M): #puts it all together

    M1 = column_stochastic_inator(M)

    M2= Google_matrix_inator(M1, x=.85)

    M3 = power_method_inator(M2, 100)

    ranks = np.argsort(M3)[::-1]

    return ranks
```

```python
#EXAMPLE WITH RANDOM WEBSITES

n = 50

M = np.random.randint(0, 2, size=(n, n)) #creates a random 50 website simulation with random links between websites

PageRank(M)

print(PageRank(M)) #the list is in order of most important ot least important where the number is the indexx for the website
```
